create a bar chart or histogram to visualize the distributuin of a categorical or continuos variable,such as the distribution of ages or genders ina population.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Load the CSV file
file_path = '/content/API_SP.POP.TOTL_DS2_en_csv_v2_2667844.csv'
df = pd.read_csv(file_path, skiprows=4)
```

```python
df.head()
```

```python

# Select data for the year 2020
data_2020 = df[['Country Name', '2020']].dropna()
data_2020.columns = ['Country', 'Population']

# Select a few countries for the bar chart
selected_countries = data_2020[data_2020['Country'].isin(['United States', 'China', 'India', 'Brazil', 'Russia'])]

```

```python

# Plotting the bar chart
plt.figure(figsize=(12, 6))
plt.bar(selected_countries['Country'], selected_countries['Population'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Total Population in 2020')
plt.title('Total Population by Country in 2020')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting the histogram for population distribution
plt.figure(figsize=(12, 6))
plt.hist(data_2020['Population'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.title('Population Distribution of Countries in 2020')
plt.show()
```

```python
# Population distribution for African countries
african_countries = df[df['Country Name'].str.contains('Africa', na=False)]
african_population = african_countries[['Country Name', '2020']].dropna()
african_population.columns = ['Country', 'Population']

plt.figure(figsize=(12, 6))
plt.hist(african_population['Population'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Population')
plt.ylabel('Number of African Countries')
plt.title('Population Distribution of African Countries in 2020')
plt.show()
```
